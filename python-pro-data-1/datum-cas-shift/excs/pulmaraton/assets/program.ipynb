```python
import pandas as pd

# Load the CSV file to check its structure
file_path = 'half_marathon.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataframe to understand its structure
data.head()
```

```python
# Filter data for 2019 and 2020
data_2019 = data[data['Rok zavodu'] == 2019]
data_2020 = data[data['Rok zavodu'] == 2020]

# Find common runners in both years by their names
common_runners = pd.merge(data_2019[['Jmeno']], data_2020[['Jmeno']], on='Jmeno')

# Extract records of common runners from the original dataset
common_runners_data = data[data['Jmeno'].isin(common_runners['Jmeno'])]

# Display the result
common_runners_data= common_runners_data.sort_values(["Jmeno", "Rok zavodu"])
common_runners_data
```

```python
common_runners_data.to_csv("half_marathon.csv", index=False)
```
